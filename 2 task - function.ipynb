```python
import numpy as np
from scipy import linalg
from matplotlib import pylab as plt

def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

```

```python
#1 степень
A = np.array([[1, 1], [15, 1]])
b= np.array([f(1), f(15)])

par = linalg.solve(A, b)

x = np.arange(1, 15, 0.1)
plt.plot(x, par[0]*x+par[1])
plt.plot(x, f(x))
plt.show()
```

```python
#2 степень
A = np.array([[1, 1, 1], [64, 8, 1],[ 225, 15, 1]])
b= np.array([f(1),f(8), f(15)])

par = linalg.solve(A, b)

x = np.arange(1, 15, 0.1)
plt.plot(x, par[0]*(x**2)+par[1]*x+par[2])
plt.plot(x, f(x))
plt.show()
```

```python
#3 степень
A = np.array([[1, 1, 1,1], [64, 16, 4, 1],[1000, 100, 10, 1], [ 15*225, 225, 15, 1]])
b= np.array([f(1),f(4),f(10), f(15)])

par = linalg.solve(A, b)

x = np.arange(1, 15, 0.1)
plt.plot(x, par[0]*(x**3)+par[1]*(x**2)+par[2]*x+par[3])
plt.plot(x, f(x))
plt.show()
```
