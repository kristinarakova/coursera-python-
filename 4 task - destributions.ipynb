```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#генерируем выборку эксп распр 
lambd = 3.0
exp_dist= sts.expon(loc=lambd)
sample=exp_dist.rvs(size=1000)
```

```python
#функция плотности распределения
x = np.linspace(0,20,100)
pdf = exp_dist.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')

```

```python
#гистограмма выборки

plt.hist(sample, bins=18, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

# на том же графике построим теоретическую плотность распределения:
x = np.linspace(0,10,100)
pdf = exp_dist.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
#генеририруем 1000 выборок размера n
n=5
matrix_distr = np.zeros((1000,n+1))
for i in range(1000):
    lambd = 3.0
    exp_dist= sts.expon()
    sample=exp_dist.rvs(n)
    matrix_distr[i,:-1] = sample
    
# подсчет выборочного среднего
matrix_distr[:,n]=matrix_distr[:,:-1].mean(axis=1)     
plot_5 = matrix_distr[:,n]
```

```python
#генеририруем 1000 выборок размера n
n=30
matrix_distr = np.zeros((1000,n+1))
for i in range(0,1000):
    lambd = 3.0
    exp_dist= sts.expon()
    sample=exp_dist.rvs(n)
    matrix_distr[i,:-1] = sample
    
# подсчет выборочного среднего
matrix_distr[:,n]=matrix_distr[:,:-1].mean(axis=1)     
plot_30 = matrix_distr[:,n]
```

```python
#генеририруем 1000 выборок размера n
n=10000
matrix_distr = np.zeros((1000,n+1))
for i in range(0,1000):
    lambd = 3.0
    exp_dist= sts.expon()
    sample=exp_dist.rvs(n)
    matrix_distr[i,:-1] = sample
    
# подсчет выборочного среднего
matrix_distr[:,n]=matrix_distr[:,:-1].mean(axis=1)     
plot_100 = matrix_distr[:,n]
```

```python
#гистограмма выборки выборочного среднего
fig=plt.figure(figsize=(16,5))
ax=fig.add_subplot(1,3,1)
ax.hist(plot_5, bins=5, normed=True)
ax.set_title('n=5', fontsize=20)
ax.set_ylabel('fraction of samples')
ax.set_xlabel('$x$')
#нормальное распределение
norm_rv = sts.norm(1, 1/np.sqrt(5))
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
ax.plot(x, pdf, label='theoretical pdf')

ax=fig.add_subplot(1,3,2)
ax.hist(plot_30, bins=30, normed=True)
ax.set_title('n=30', fontsize=20)
ax.set_ylabel('fraction of samples')
ax.set_xlabel('$x$')
#нормальное распределение
norm_rv = sts.norm(1,1/np.sqrt(30))
x = np.linspace(0,2,100)
pdf = norm_rv.pdf(x)
ax.plot(x, pdf, label='theoretical pdf')


ax=fig.add_subplot(1,3,3)
ax.hist(plot_100,bins=500,  normed=True)
ax.set_title('n=10000', fontsize=20)
ax.set_ylabel('fraction of samples')
ax.set_xlabel('$x$')
#нормальное распределение
norm_rv = sts.norm(1, 1/np.sqrt(10000))
x = np.linspace(0.95,1.05,100)
pdf = norm_rv.pdf(x)
ax.plot(x, pdf, label='theoretical pdf');
```
